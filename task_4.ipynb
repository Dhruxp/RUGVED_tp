# Linear Regression

```python
import numpy as np
# X - Engine Size, Fuel Consumption, Cylinders
# y - CO2 Emissions
X = df[["Engine Size(L)", "Fuel Consumption Comb (L/100 km)", "Cylinders"]].values
y = df["CO2 Emissions(g/km)"].values
```

```python
# Standardize the features
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X - X_mean) / X_std # Scaling so that values are comparable
```

```python
#parameterization
m, n = X.shape
w = np.zeros(n)
b = 0.0
# weights and bias 
```

```python
def predict(X, w, b):
    return np.dot(X, w) + b
# pred
```

```python
def compute_cost(X, y, w, b):
    h = len(y)
    y_pred = predict(X, w, b)
    return (1 / h) * np.sum((y_pred - y) ** 2)
# loss calc 
```

```python
def gradient_descent(X, y, w, b, lr=0.01, epochs=1000):
    m = len(y)
    cost_history = []
    for epoch in range(epochs):
        y_pred = predict(X, w, b)
        dw = (1 / m) * np.dot(X.T, (y_pred - y))
        db = (1 / m) * np.sum(y_pred - y)
        w -= lr * dw
        b -= lr * db
        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)
        if epoch % 100 == 0:
            cost = compute_cost(X, y, w, b)
            print(f"Epoch {epoch}, Cost: {cost:.4f}")
    return w, b, cost_history
#engine
```

```python
w, b, cost_history = gradient_descent(
    X, y, w, b,
    lr=0.01,
    epochs=1000
)
print("\nTrained Weights:", w)
print("Trained Bias:", b)
```

```python
#curve - loss fn
plt.plot(cost_history)
plt.xlabel("Epochs")
plt.ylabel("Cost")
```

```python
#eval
r2 = 1 - (np.sum((predict(X, w, b) - y) ** 2) / np.sum((y - np.mean(y)) ** 2))
print(f"R2 Score: {r2:.4f}")
```

```python
y_pred = predict(X, w, b)
plt.figure()
plt.scatter(y, y_pred)
plt.xlabel("Actual CO2 Emissions")
plt.ylabel("Predicted CO2 Emissions")
plt.title("Actual vs Predicted CO2 Emissions")
plt.show()
```
