```python
# Write a python program to check if given number is a hill number.
def hill_num(num):
    num_str = str(num)
    length = len(num_str)
    if length < 3:
        return False
    peak_found = False
    for i in range(1, length):
        if not peak_found:
            if num_str[i] >= num_str[i - 1]:
                continue
            elif num_str[i] < num_str[i - 1]:
                peak_found = True
            else:
                return False
        else:
            if num_str[i] < num_str[i - 1]:
                continue
            else:
                return False
    return peak_found
resfin = hill_num(1234421)
resfin
```

```python
# Write a python function to perform selection sort on a given string.
def selection_sort_string(s):
    s_list = list(s)
    n = len(s_list)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if s_list[j] < s_list[min_index]:
                min_index = j
        s_list[i], s_list[min_index] = s_list[min_index], s_list[i]
    return ''.join(s_list)
sorted_string = selection_sort_string("python")
sorted_string
```

```python
# Find the factorial of a given number using recursion.
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
fc = factorial(5)
fc
```

```python
# Create a function that checks whether given string is an anagram or not?
def check_anagrams(str1, str2):
    str1_sorted = sorted(str1)
    str2_sorted = sorted(str2)
    return str1_sorted == str2_sorted
testfunc = check_anagrams("listen", "silent")
testfunc
```

```python
# Write a program to print the Fibonacci Sequence till n-values where n is user input.
def fibonacci_sequence(n):
    fib_seq = []
    a, b = 0, 1
    while a < n:
        fib_seq.append(a)
        a, b = b, a + b
    return fib_seq
n = int(input("Enter a number: "))
fib_seq = fibonacci_sequence(n)
print(fib_seq)
```

```python
# Write a python program to divide a given string into equal parts containing n (user input) characters in each part.
def divide_string(s, n):
    if len(s) % n != 0:
        return "Undivisible"
    return [s[i:i + n] for i in range(0, len(s), n)]
stringneeded = input("Enter a string: ")
partsize = int(input("Enter the size of each part: "))
divided_parts = divide_string(stringneeded, partsize)
print(divided_parts)
```

```python
# Write a Python function to encrypt a string using Caesar’s Cipher
def caesar_cipher(text, shift):
    encrypted_text = ""
    for char in text:
        if char.isalpha():
            shift_base = ord('A') if char.isupper() else ord('a')
            encrypted_char = chr((ord(char) - shift_base + shift) % 26 + shift_base)
            encrypted_text += encrypted_char
        else:
            encrypted_text += char
    return encrypted_text
txt = input("Enter a string to encrypt: ")
print(txt)
shft = int(input("Enter shift value: "))
encrypted = caesar_cipher(txt, shft)
print(encrypted)
```
